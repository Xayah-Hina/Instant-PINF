```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# 创建 3D 图形
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# 第一组网格 (3 x 3 x 3)
x1 = np.linspace(-5, 5, 3)
y1 = np.linspace(-5, 5, 3)
z1 = np.linspace(-5, 5, 3)

# 第二组网格 (4 x 4 x 4)
x2 = np.linspace(-5, 5, 5)
y2 = np.linspace(-5, 5, 5)
z2 = np.linspace(-5, 5, 5)

# 创建第一组网格点
X1, Y1 = np.meshgrid(x1, y1)
X2, Y2 = np.meshgrid(x2, y2)

# 使用 z 的网格点生成第一组网格线（实线）
for z_level in z1:
    ax.plot_wireframe(X1, Y1, z_level * np.ones_like(X1), color='black', linestyle='-', linewidth=2, alpha=0.8)  # 浅紫色

# 使用 z 的网格点生成第二组网格线（虚线）
for z_level in z2:
    ax.plot_wireframe(X2, Y2, z_level * np.ones_like(X2), color='gray', linestyle='--', linewidth=1.5, alpha=0.8)  # 浅橙色

# 添加第一组网格的坐标轴线条（实线）
for x_line in x1:
    for y_line in y1:
        ax.plot([x_line, x_line], [y_line, y_line], [-5, 5], color='black', linestyle='-', linewidth=2, alpha=0.8)

# 添加第二组网格的坐标轴线条（虚线）
for x_line in x2:
    for y_line in y2:
        ax.plot([x_line, x_line], [y_line, y_line], [-5, 5], color='gray', linestyle='--', linewidth=1.5, alpha=0.8)

# 设置图形外观
ax.axis('off')  # 关闭坐标轴
ax.set_xlim([-5, 5])
ax.set_ylim([-5, 5])
ax.set_zlim([-5, 5])
ax.view_init(elev=20, azim=30)  # 设置视角
ax.grid(False)  # 关闭默认网格
ax.set_box_aspect([1, 1, 1])  # 均匀比例

# 保存为高分辨率图片
plt.savefig('3d_grids_combined_low_saturation.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# 第二组网格 (5 x 5 x 5)
x2 = np.linspace(-5, 5, 5)
y2 = np.linspace(-5, 5, 5)
z2 = np.linspace(-5, 5, 5)

# 创建网格点
X2, Y2 = np.meshgrid(x2, y2)

# 绘制第二组网格
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# 绘制最外层立方体的8条边（实线）
for z_level in [z2[0], z2[-1]]:  # 顶部和底部的 z 平面
    for x in [x2[0], x2[-1]]:  # x 平面边缘
        ax.plot([x, x], [y2[0], y2[-1]], [z_level, z_level], color='black', linestyle='-', linewidth=1.5, alpha=0.8)  # 垂直边
    for y in [y2[0], y2[-1]]:  # y 平面边缘
        ax.plot([x2[0], x2[-1]], [y, y], [z_level, z_level], color='black', linestyle='-', linewidth=1.5, alpha=0.8)  # 水平边

for x in [x2[0], x2[-1]]:  # 竖直边连接顶部和底部
    for y in [y2[0], y2[-1]]:
        ax.plot([x, x], [y, y], [z2[0], z2[-1]], color='black', linestyle='-', linewidth=1.5, alpha=0.8)

# 绘制内部虚线网格
for z_level in z2:
    ax.plot_wireframe(X2, Y2, z_level * np.ones_like(X2), color='black', linestyle='--', linewidth=0.8, alpha=0.5)
for x_line in x2:
    for y_line in y2:
        if not (x_line in [x2[0], x2[-1]] and y_line in [y2[0], y2[-1]]):  # 排除实线边
            ax.plot([x_line, x_line], [y_line, y_line], [z2[0], z2[-1]], color='black', linestyle='--', linewidth=0.8, alpha=0.5)

# 设置图形外观
ax.axis('off')
ax.set_xlim([-5, 5])
ax.set_ylim([-5, 5])
ax.set_zlim([-5, 5])
ax.view_init(elev=20, azim=30)
ax.set_box_aspect([1, 1, 1])
plt.savefig('3d_cube_edges_solid_rest_dashed.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 创建 2D 图形
fig, ax = plt.subplots(figsize=(8, 6))

# 第一组网格 (3 x 3)
x1 = np.linspace(-5, 5, 3)
y1 = np.linspace(-5, 5, 3)

# 第二组网格 (5 x 5)
x2 = np.linspace(-5, 5, 4)
y2 = np.linspace(-5, 5, 4)

# 第一组网格（实线）
for x_line in x1:
    ax.plot([x_line, x_line], [y1[0], y1[-1]], color='gray', linestyle='-', linewidth=2, alpha=0.7)  # 垂直线
for y_line in y1:
    ax.plot([x1[0], x1[-1]], [y_line, y_line], color='gray', linestyle='-', linewidth=2, alpha=0.7)  # 水平线

# 第二组网格（虚线）
for x_line in x2:
    ax.plot([x_line, x_line], [y2[0], y2[-1]], color='gray', linestyle='--', linewidth=1.5, alpha=0.7)  # 垂直线
for y_line in y2:
    ax.plot([x2[0], x2[-1]], [y_line, y_line], color='gray', linestyle='--', linewidth=1.5, alpha=0.7)  # 水平线

# 设置图形外观
ax.set_xlim([-5, 5])
ax.set_ylim([-5, 5])
ax.set_aspect('equal', adjustable='box')  # 确保网格比例一致
ax.axis('off')  # 关闭坐标轴

# 保存为高分辨率图片
plt.savefig('2d_grids_combined.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 定义网格的已知点坐标 (四个角)
x = [0, 1]
y = [0, 1]

# 定义四个已知点的值
f = np.array([[1, 2],  # f(0,0)=1, f(1,0)=2
              [3, 4]])  # f(0,1)=3, f(1,1)=4

# 定义插值点的坐标
x_interp, y_interp = 0.3, 0.7

# 计算插值点的值（基于双线性插值公式）
f_interp = (f[0, 0] * (1 - x_interp) * (1 - y_interp) +
            f[1, 0] * x_interp * (1 - y_interp) +
            f[0, 1] * (1 - x_interp) * y_interp +
            f[1, 1] * x_interp * y_interp)

# 绘制网格
fig, ax = plt.subplots(figsize=(6, 6))

# 绘制已知点的网格
ax.scatter([0, 1, 0, 1], [0, 0, 1, 1], color='blue', s=100, label='Grid Points (Known)')
for i, txt in enumerate(['f(0,0)=1', 'f(1,0)=2', 'f(0,1)=3', 'f(1,1)=4']):
    ax.annotate(txt, ([0, 1, 0, 1][i], [0, 0, 1, 1][i]), textcoords="offset points", xytext=(0, 5), ha='center', fontsize=10)

# 绘制插值点
ax.scatter(x_interp, y_interp, color='red', s=100, label=f'Interpolation Point\nf({x_interp},{y_interp})={f_interp:.2f}')
ax.annotate(f'f({x_interp:.1f},{y_interp:.1f})={f_interp:.2f}', (x_interp, y_interp), textcoords="offset points",
            xytext=(-70, -15), ha='center', fontsize=10, color='red')

# 绘制矩形框
ax.plot([0, 1, 1, 0, 0], [0, 0, 1, 1, 0], color='black', linestyle='--', label='Grid Boundary')

# 设置坐标轴范围和标题
ax.set_xlim(-0.2, 1.2)
ax.set_ylim(-0.2, 1.2)
ax.set_aspect('equal')
ax.set_title('Bilinear Interpolation Illustration', fontsize=14)
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('y', fontsize=12)

# 添加图例
ax.legend(loc='upper left', fontsize=10)

# 显示图形
plt.grid(True, linestyle='--', alpha=0.5)
plt.savefig('bilinear_interpolation_demo.png', dpi=300, bbox_inches='tight')
plt.show()
```
