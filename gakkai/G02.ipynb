```python
import numpy as np
import os
```

```python
sim_id = 15
sim_dir = os.path.normpath(os.path.join(
    'C:/Users/imeho/Documents/ScalarFlow',
    f"sim_{sim_id:06d}",
))
```

```python
import matplotlib.pyplot as plt

frame_id = 30
file_name = f"imgsTarget_{frame_id:06d}.npz"
file_path = os.path.join(sim_dir, file_name)
data = np.load(file_path)['data']

fig, axes = plt.subplots(1, 5, figsize=(20, 10))  # 创建1行5列的子图
for i in range(5):
    axes[i].imshow(np.flipud(data[i]), cmap='gray')
    axes[i].set_title(f"Image {i + 1}")
    axes[i].axis('off')
plt.tight_layout()
plt.show()
```
