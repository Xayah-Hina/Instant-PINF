```python
import numpy as np
import matplotlib.pyplot as plt

# 生成速度场数据
x, y = np.meshgrid(np.linspace(-1, 1, 50), np.linspace(-1, 1, 50))
u = -y
v = x
magnitude = np.sqrt(u**2 + v**2)

# 绘制速度大小等高线
plt.figure(figsize=(6, 6))
plt.streamplot(x, y, u, v, color=magnitude, cmap='viridis')
plt.colorbar(label='Velocity Magnitude')
plt.title('2D Slice of 3D Velocity Field')
plt.xlabel('X')
plt.ylabel('Y')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# 创建 3D 图形
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# 设置网格范围
x = np.linspace(-5, 5, 4)
y = np.linspace(-5, 5, 4)
z = np.linspace(-5, 5, 4)

# 创建网格点
X, Y = np.meshgrid(x, y)

# 使用 z 的网格点生成网格线
for z_level in z:
    ax.plot_wireframe(X, Y, z_level * np.ones_like(X), color='gray', linewidth=0.5, alpha=0.7)

# 添加坐标轴线条
for x_line in x:
    for y_line in y:
        ax.plot([x_line, x_line], [y_line, y_line], [-5, 5], color='gray', linewidth=0.5, alpha=0.7)

# 设置坐标轴标签
ax.set_xlabel('X-axis', labelpad=10, fontsize=12)
ax.set_ylabel('Y-axis', labelpad=10, fontsize=12)
ax.set_zlabel('Z-axis', labelpad=10, fontsize=12)

# 设置坐标轴范围
ax.set_xlim([-5, 5])
ax.set_ylim([-5, 5])
ax.set_zlim([-5, 5])

# 调整图形外观
ax.view_init(elev=20, azim=30)  # 设置视角
ax.grid(False)  # 关闭默认网格
ax.set_box_aspect([1, 1, 1])  # 均匀比例

# 保存为高分辨率图片
plt.savefig('3d_grids.png', dpi=300, bbox_inches='tight')
plt.show()
```
