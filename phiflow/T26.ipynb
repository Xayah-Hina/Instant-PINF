```python
from phi.torch.flow import *
import tqdm

device = torch.device('cuda')
BOUNDS = Box(x=1, y=1)
RESOLUTION = 100


@jit_compile
def step(v, s, dt, inflow):
    s = advect.mac_cormack(s, v, dt) + 0.2 * resample(inflow, to=s, soft=True)
    buoyancy = resample(s * (0, 0.1), to=v)
    v = advect.semi_lagrangian(v, v, dt) + buoyancy * dt
    v, _ = fluid.make_incompressible(v, (), Solve('auto', 1e-3))
    return v, s


v0 = StaggeredGrid(
    values=0,
    extrapolation=extrapolation.ZERO,
    bounds=BOUNDS,
    resolution=spatial(x=RESOLUTION, y=RESOLUTION),
)
smoke0 = CenteredGrid(
    values=0,
    extrapolation=extrapolation.ZERO_GRADIENT,
    bounds=BOUNDS,
    resolution=spatial(x=RESOLUTION, y=RESOLUTION),
)
inflow0 = Sphere(x=0.5, y=0.15, radius=0.1)

vel, smoke = v0, smoke0
vel_list = []
smoke_list = []
for _ in tqdm.trange(100):
    vel, smoke = step(vel, smoke, 0.2, inflow0)
    vel_list.append(vel)
    smoke_list.append(smoke)
```

```python
keyframes = 5
sim_start = 20
sim_iters = 3
smoke_record_list = []
for _ in range(keyframes):
    smoke_record_list.append(smoke_list[sim_start + sim_iters * _])

_ = plot(*smoke_record_list[:keyframes // 2])
_ = plot(*smoke_record_list[keyframes // 2:])
```

```python
def step(v, s, fr, dt, ir, inflow):
    s = advect.mac_cormack(s, v, dt) + ir * resample(inflow, to=s, soft=True)
    v = advect.semi_lagrangian(v, v, dt) + dt * fr
    return v, s


@jit_compile
def loss_function(in_vel, in_den, in_force_list, in_dt, in_inflow_rate, in_inflow):
    iter_v = in_vel
    iter_s = in_den
    losses = []
    for force_idx in range(len(in_force_list)):
        for _ in range(sim_iters):
            iter_v, iter_s = step(iter_v, iter_s, in_force_list[force_idx], in_dt, in_inflow_rate, in_inflow)
        losses.append(math.l2_loss(iter_s.values - smoke_record_list[force_idx].values))
    return sum(losses)


grad_fun = math.gradient(loss_function, wrt='in_vel, in_force_list', get_output=True)


def gradient_descent_step(in_vel, in_den, in_force_list, in_dt, in_inflow_rate, in_inflow, learning_rate=.0001):
    out_loss, (d_vel, d_frs) = grad_fun(in_vel, in_den, in_force_list, in_dt, in_inflow_rate, in_inflow)
    return out_loss, in_vel - learning_rate * d_vel, {key: in_force_list[key] - 0.05 * learning_rate * d_frs[key] for
                                                      key in range(len(in_force_list))}
```

```python
velocity_to_train = StaggeredGrid(
    values=0,
    extrapolation=extrapolation.ZERO_GRADIENT,
    bounds=BOUNDS,
    resolution=spatial(x=RESOLUTION, y=RESOLUTION),
)
force_to_train_list = []
for _ in range(keyframes - 1):
    force_to_train_list.append(StaggeredGrid(
        values=0,
        extrapolation=extrapolation.ZERO_GRADIENT,
        bounds=BOUNDS,
        resolution=spatial(x=RESOLUTION, y=RESOLUTION),
    ))
losses = []
for _ in tqdm.trange(100):
    loss, velocity_to_train, force_to_train_list = gradient_descent_step(velocity_to_train, smoke_record_list[0],
                                                                                      force_to_train_list, 0.2, 0.2, inflow0)
    losses.append(loss)
```

```python
import matplotlib.pyplot as plt

# 绘制损失曲线
plt.figure(figsize=(10, 6))
plt.plot(losses, label="Training Loss", linestyle='-', marker='o')  # 使用线条和标记
plt.xlabel("Epochs")  # 横轴标签
plt.ylabel("Loss")  # 纵轴标签
plt.title("Loss Curve")  # 图标题
plt.legend()  # 显示图例
plt.grid(True)  # 添加网格
plt.show()
```
